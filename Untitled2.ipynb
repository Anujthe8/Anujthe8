```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import PCA
import matplotlib.colors as colors
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

df = pd.read_csv('/content/12.csv', header=1)

# Explore the first ten rows of the data set
df.head(10)

# Filter the DataFrame
df_no_missing_data = df.loc[(df['EDUCATION'] != 0) & (df['MARRIAGE'] != 0)]

# Verify the column names in the DataFrame
print(df_no_missing_data.columns)

# Assuming the correct column name is 'default payment next month' based on the dataset
# count plot on output variable
ax = sns.countplot(x=df_no_missing_data['default payment next month'], hue=df_no_missing_data['default payment next month'], palette='rocket', legend=False)

#add data labels
ax.bar_label(ax.containers[0])

# add plot title
plt.title("Observations by Classification Type")

# show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Reset index to make 'DEFAULT' a regular column if it's currently an index
df_no_missing_data = df_no_missing_data.reset_index()

# Now you can pivot and plot as intended
df_no_missing_data.pivot_table(index='MARRIAGE', columns='default payment next month', values='AGE').plot(kind='bar')
plt.xlabel('Marriage')
plt.ylabel('Age')
plt.title('Default Payment by Marriage and Age')
plt.show()
```

```python
# prompt: Draw a bar chart for default payment  including Age field

import matplotlib.pyplot as plt
# Create a new DataFrame with the desired columns
df_chart = df_no_missing_data[['default payment next month', 'MARRIAGE', 'AGE']]

# Use Seaborn to create a countplot with multiple hues
sns.countplot(x="default payment next month", hue="AGE", data=df_chart, palette="rocket")

# Add a title and labels
plt.title("Default Payment by Age")
plt.xlabel("Default")
plt.ylabel("Count")

# Show the plot
plt.show()
```

```python
# prompt: Draw a KS chart for evaluating which customers have default enable in their credit card payments

import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

# Define the event and time columns
event_col = "default payment next month"
time_col = "AGE"

# Create a Kaplan-Meier estimator
kmf = KaplanMeierFitter()

# Fit the estimator to the data
kmf.fit(df_no_missing_data[time_col], event_observed=df_no_missing_data[event_col])

# Plot the Kaplan-Meier curve
kmf.plot()

# Add a title and labels
plt.title("Kaplan-Meier Curve for Default Payment")
plt.xlabel("Age")
plt.ylabel("Probability of Default")

# Show the plot
plt.show()
```
